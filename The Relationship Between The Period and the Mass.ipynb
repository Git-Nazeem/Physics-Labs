# The Relationship between The Time of Oscillation and the Mass

## Introduction

An Oscillation is a back and forth repetitive motion. The object in motion will return to its original position after displacement. A Period is the time it takes for one oscillation to occur.

## Procedure

* Gather the materials: spring and weights (50g, 100g, 150g, and 200g).
* Place a weight on the plate of the spring and then record the amount of oscillations within a 30 second period, for each weight.
* Repeat the procedures for each weight and record the data.

## Data:

| Mass(g) | Number of Oscillations |
|---|---|
|$$50g$$|$$90$$|
|$$100g$$|$$79$$|
|$$150g$$|$$56$$|
|$$200g$$|$$50$$|

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

mass = np.array([50, 100, 150, 200])
osc = np.array([90, 79, 56, 50])
xx = np.linspace(0.0, 200, 20)

def lin_model(x,a,b):
    return a*x + b

a,b = curve_fit(lin_model, mass, osc)[0]
print=(a,b)

plt.plot(mass, osc, '.r', label = 'Relationship')
plt.legend(loc='upper right')
plt.xlabel('Mass (g)')
plt.ylabel('Number of Oscillations in 30 Sec')
plt.title('Relationship Between the Number of Oscillations and Time')
plt.plot(xx, lin_model(xx,a,b))
plt.show
```

## Results and Observations

After we reviewed the video and took count of the ammount of ossolations that occured in the 30 second time frame, we noticed that the more weight we put into the spring the less ossoillations there is. We can say that this is happening because the more weight there is in the spring, the more harder it is for the spring to go back up since the pull back up is being distrubted by the weights.
